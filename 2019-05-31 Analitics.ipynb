```python
years = np.arange(2011,2019)
years
```

```python
spam = []
for i in range(0,8):
    spam.append(int(data.loc[data['год'] == years[i], 'НМЦК'].sum()))

statistics = pd.DataFrame(spam, index = years, columns = ['общий объём закупок'])
# , 'из них РСИ','из них мед.отходы','средняя НМЦК'
statistics
```

```python
statistics['из них РСИ'] = np.round(data.loc[(data['Тип закупки'] == 'Ртуть'), ['НМЦК','год']].groupby('год')[['НМЦК']].sum())
statistics
```

```python
statistics['из них мед.отходы'] = np.round(data.loc[(data['Тип закупки'] == 'Мед.отходы'), ['НМЦК','год']].groupby('год')[['НМЦК']].sum())
statistics
```

```python
statistics['средняя НМЦК'] = np.round(data.groupby('год')[['НМЦК']].mean(), 0)
statistics
```

```python
statistics['общее кол-во процедур'] = np.round(statistics['общий объём закупок'] / statistics['средняя НМЦК'], 0)
statistics
```

```python
statistics['экономия %'] = np.round(data.groupby('год')[['Экономия в %']].mean(), 0)
statistics
```

```python
spam = []

smp = np.round(data.loc[(data['год'] == 2019)].groupby('Ограничения')[['НМЦК']].sum(), 0)
smp.loc['Всего'] = np.round(data.loc[(data['год'] == 2019), ['НМЦК']].sum(), 0)

spam = smp.loc['Всего'].values
smp['%'] = np.round(smp['НМЦК'] * 100 / spam, 0)

smp
```
